### AttnGAN training on filtered captions
Here, I will train the AttnGAN on only 1 caption, which has been filtered on word-frequency and named entities are replaced by 'person'/'location'/'number', etc. 

**!!!**
- in `datasets.py` 
  - in `load_captions` change `captions = f.read().split('\n')` to `captions = [f.read().split('\n')[2]]` in order to only train on the last, filtered caption. 
  - in function `load_filenames`, change `filepath = '%s/%s/xxx_filenames.pickle'` to `filepath = '%s/%s/filter_filenames.pickle` to prevent training on the images with the empty captions.
  - change `sent_ix = random.randint(0, (self.embeddings_num-1))` to `sent_ix = random.randint(0, (self.embeddings_num))` 

- Change this back once you're done with the filtered captions!
